```python
from plotnine import *
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
mtcars.tail()
```

```python
mtcars.head(15)
```

```python
(ggplot(mtcars,
        aes('wt','mpg'))
+ geom_point()
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('mpg','wt'))
+geom_point()
+theme_classic()
)
```

```python
(ggplot(mtcars,
        aes('mpg','wt'))
+geom_point()
)
```

```python
(ggplot(mtcars,
        aes('wt','mpg',color='factor(gear)'))
+ geom_point()
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('mpg','wt', color='gear'))
+geom_point()
+theme_classic())
```

```python
(ggplot(mtcars,
        aes('wt','mpg',color='factor(gear)',size='cyl'))
+geom_point()
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('wt','mpg', color='gear', size='cyl'))
+geom_point()
+theme_bw())
```

```python
(ggplot(mtcars,
       aes('wt', 'mpg', color = 'factor(gear)'))
+geom_point()
+facet_wrap('cyl')
+theme_bw()
)
```

```python
(ggplot(mtcars,
       aes('wt', 'mpg', color = 'factor(gear)'))
+geom_point()
+facet_wrap('~cyl')
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('wt','mpg', color='factor(gear)'))
+geom_point()
+facet_grid('~cyl')
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
+geom_point()
+stat_smooth(method='lm')
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
+geom_point()
+stat_smooth(method='lm')
+theme_classic())
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color= 'factor(gear)', size='factor(am)'))
+geom_point()
+facet_wrap('~cyl')
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
+geom_point()
+facet_wrap('~cyl')
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)', size='cyl'))
+geom_point()
+facet_wrap('~am')
+stat_smooth(method='lm')
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)',
        size='cyl'))
+geom_point()
+facet_grid('am ~ carb')
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('mpg', 'wt',
            color='factor(gear)', size='factor(am)'))
+geom_point()
+facet_grid('am ~ carb')
+theme_bw()
)
```
