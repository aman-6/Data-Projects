```python
from plotnine import *
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
mtcars.tail()
```

```python
(ggplot(mtcars,
        aes('hp', 'cyl'))
+geom_point()
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('hp', color='factor(cyl)'))
+geom_bar()
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('mpg', 'cyl', size='hp'))
+geom_point()
+theme_gray())
```

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
+geom_point()
+theme_bw()
)
```

```python
(ggplot(mtcars,
        aes('wt', 'hp', color='factor(gear)'))
+geom_point()
+facet_wrap('~cyl')
+theme_bw())
```

```python
(ggplot(mtcars,
        aes('wt','mpg', color='factor(gear)'))
+geom_point()
+theme_bw())
```

```python
(ggplot(
    mtcars,
    aes('wt', 'mpg', color='factor(gear)', size='cyl'))
    +geom_point()
    +facet_grid('~am')
    +stat_smooth(method='lm')
    +theme_bw())
```

```python
(ggplot(mtcars,
        aes('wt','mpg',color='factor(gear)',size='cyl'))
+geom_point()
+facet_grid('am~carb')
+theme_bw())
```
